**Sampling from a distribution**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Poisson Distribution with replacement and without replacement

```python
# Poisson distribution with lambda = 3
lambd = 3

# Generate random population
n_samples = 100000
population = np.random.poisson(lambd, size=n_samples)

# Generate random samples with replacement
samples_wr = np.random.choice(population, size=n_samples, replace=True)

# Generate random samples without replacement
samples_wor = np.random.choice(population, size=n_samples, replace=False)

# Plot histograms of the samples
sns.histplot(samples_wr, kde=True, label='With replacement', color="red")
sns.histplot(samples_wor, kde=True, label='Without replacement', color="green")
plt.legend()
plt.show()
```

Bernoulli Distribution with replacement and without replacement

```python
# Bernoulli distribution with probability of success = 0.3
p = 0.3

# Generate random population
n_samples = 100000
population = np.random.binomial(1, p, size=n_samples)

# Generate random samples with replacement
samples_wr = np.random.choice(population, size=n_samples, replace=True)

# Generate random samples without replacement
samples_wor = np.random.choice(population, size=n_samples, replace=False)

# Plot histograms of the samples
sns.histplot(samples_wr, kde=True, label='With replacement', color="red")
sns.histplot(samples_wor, kde=True, label='Without replacement', color="green")
plt.legend()
plt.show()
```

When we generate samples with replacement, each value has the same chance of being selected and the resulting samples have a similar distribution to the underlying one. When generating samples without replacement, the chances of selecting each value changes as the number of available values decreases. But as the sample size increases the distribution of the samples will become similar to the underlying distribution. The difference between samples generated with and without replacement will be more noticeable when the underlying distribution has high variance or skewness and if the sample size is small.
